```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
students_data = "StudentsPerformance.csv"


# Read School and Student Data File and store into Pandas Data Frames
students_data = pd.read_csv(students_data)
```

```python
students_data
```

```python
totalstudents= len(students_data['gender'].value_counts())
Male= students_data.groupby("gender").get_group('male')
totalmales= len(Male)
female= students_data.groupby("gender").get_group('female')
totalfemales=len(female)
```

```python

Male.head()
```

```python
male_math_passing =Male[(Male['math score']>=70)].count()['math score']
male_math_passing_rate=round(male_math_passing/totalmales*100,2)
male_math_passing_rate
```

```python
male_reading_passing =Male[(Male['reading score']>=70)].count()['reading score']
male_reading_passing_rate=round(male_reading_passing/totalmales*100,2)
male_writing_passing =Male[(Male['writing score']>=70)].count()['writing score']
male_writing_passing_rate=round(male_writing_passing/totalmales*100,2)
```

```python
female_math_passing=female[(female['math score']>=70)].count()['math score']
female_math_passing_rate=round(female_math_passing/totalfemales*100,2)
female_reading_passing=female[(female['reading score']>=70)].count()['reading score']
female_reading_passing_rate=round(female_reading_passing/totalfemales*100,2)
female_writing_passing=female[(female['writing score']>=70)].count()['writing score']
female_writing_passing_rate=round(female_writing_passing/totalfemales*100,2)
```

```python
Genderpurchanalysis = {"Total":[totalmales,totalfemales],
                    "Gender":["Male","Female"],
                    "Math_passing_rate":[male_math_passing_rate,female_math_passing_rate],
                    "Reading_passing_persentage":[male_reading_passing_rate,female_reading_passing_rate],
                "Writing_passing_count":[male_writing_passing_rate,female_writing_passing_rate]}
Genderpurchanalysis1= pd.DataFrame(Genderpurchanalysis)
Genderpurchanalysis1 = Genderpurchanalysis1.set_index('Gender')
Genderpurchanalysis1= Genderpurchanalysis1[['Total','Math_passing_rate','Reading_passing_persentage','Writing_passing_count']]
Genderpurchanalysis1
```
